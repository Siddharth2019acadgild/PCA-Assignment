```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris() # import dataset
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
X = iris.data
y = iris.target 
```

```python
fig = plt.figure( figsize=(4, 3)) # Making axes 3Dimmensional 
plt.show()
ax = Axes3D(fig)
```

```python
fig
```

```python
pca = decomposition.PCA(n_components=3) # Implementing PCA
pca.fit(X)
X = pca.transform(X)
```

```python
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, edgecolor='k') # Plotting the scatter plot.
```
